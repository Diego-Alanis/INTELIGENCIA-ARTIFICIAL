```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
url='https://raw.githubusercontent.com/dilp90/InteligenciaArtificial_y_RedesNeuronales_UANL_FIME/main/MachineLearning/Datasets/robot_sensors.csv'
data=pd.read_csv(url)
data.head()
```

```python
data.info()
```

```python
data_clean=data.dropna()
data_clean=data_clean[data_clean['Velocidad (cm/s)']>=0]
data_clean.head()
```

```python
data_norm=(data_clean-data_clean.min())/(data_clean.max()-data_clean.min())
data_norm.head()
```

```python
plt.figure(figsize=(10,6))
plt.plot(data_norm['Tiempo (s)'],data_norm['Distancia (cm)'],label='Distancia')
plt.plot(data_norm['Tiempo (s)'],data_norm['Velocidad (cm/s)'],label='Velocidad')
plt.plot(data_norm['Tiempo (s)'],data_norm['Temperatura (°C)'],label='Temperatura')
plt.xlabel('Tiempo')
plt.ylabel('Valores normalizados')
plt.title('Datos normalizados de sensores del robot')
plt.legend()
plt.show()
```
