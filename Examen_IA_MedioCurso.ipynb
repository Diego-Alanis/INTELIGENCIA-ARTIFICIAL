# Parte 1: Carga y Limpieza de Datos

En esta sección se realiza la carga del dataset desde una fuente externa, se inspeccionan los datos, se identifican valores nulos y se realiza la limpieza correspondiente.  
Además, se genera una nueva variable para mejorar el análisis de los datos.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/mpg.csv"
df = pd.read_csv(url)

# Mostrar primeras 5 filas
df.head()
```

```python
df.dtypes
```

## Limpieza de Datos

Se identificaron valores nulos en la columna *horsepower*.  
Para mantener la integridad del análisis, estos valores fueron reemplazados utilizando la mediana de la columna, evitando así distorsiones en los datos.

```python
# Ver valores nulos en horsepower
df['horsepower'].isnull().sum()

# Reemplazar nulos con la mediana
mediana_hp = np.nanmedian(df['horsepower'])
df['horsepower'] = df['horsepower'].fillna(mediana_hp)
```

## Creación de Nueva Variable

Se creó una nueva columna llamada *relacion_peso_potencia*, la cual permite analizar la relación entre el peso del vehículo y su potencia, proporcionando una métrica adicional para el estudio del rendimiento.

```python
df['relacion_peso_potencia'] = df['weight'] / df['horsepower']

df.head()
```

# Parte 2: Visualización de Datos (Matplotlib)

En esta sección se generan distintas gráficas para analizar el comportamiento de los datos del dataset Auto MPG.

## Histograma de Aceleración

Esta gráfica muestra la distribución de la aceleración de los vehículos, permitiendo observar la frecuencia de los valores en diferentes rangos.

```python
plt.figure()
plt.hist(df['acceleration'])
plt.title("Distribución de la Aceleración")
plt.xlabel("Aceleración")
plt.ylabel("Frecuencia")
plt.show()
```

## Gráfico de Dispersión

Se analiza la relación entre el desplazamiento del motor (displacement) y los caballos de fuerza (horsepower), identificando posibles patrones o tendencias.

```python
plt.figure()
plt.scatter(df['displacement'], df['horsepower'])
plt.title("Displacement vs Horsepower")
plt.xlabel("Desplazamiento")
plt.ylabel("Caballos de fuerza")
plt.show()
```

## Gráfico de Barras

Se muestra el promedio de rendimiento (mpg) agrupado por el país de origen, lo que permite comparar el desempeño entre diferentes regiones.

```python
# Promedio de mpg por origen
promedio_mpg = df.groupby('origin')['mpg'].mean()

plt.figure()
promedio_mpg.plot(kind='bar')
plt.title("Promedio de MPG por País de Origen")
plt.xlabel("Origen")
plt.ylabel("MPG promedio")
plt.show()
```
